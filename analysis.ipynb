```python
import pandas as pd

# clean columns
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")

# convert to numeric (VERY IMPORTANT)
df['math_score'] = pd.to_numeric(df['math_score'], errors='coerce')
df['reading_score'] = pd.to_numeric(df['reading_score'], errors='coerce')
df['writing_score'] = pd.to_numeric(df['writing_score'], errors='coerce')

# create total and avg score
df['total_score'] = df[['math_score','reading_score','writing_score']].sum(axis=1)
df['avg_score'] = df['total_score'] / 3

# result
print(df.groupby('gender')['avg_score'].mean())
```
